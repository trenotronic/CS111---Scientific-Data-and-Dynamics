```python
def splitIntoSentences(text):
    endPunc = ".!?"    # Initialize variables
    sentence = ''
    sentenceList = []
    prevChar = ""  # Initialize as an empty string
    for char in text:
        if char != " ":
            sentence += char
        elif prevChar in endPunc:
            if sentence:
                sentenceList.append(sentence)
                sentence = ''
        prevChar = char
    if sentence:
        sentenceList.append(sentence)
    return sentenceList
def main():
    s = " I really hope this works. Who knows? Let's see!"
    print(splitIntoSentences(s))
main()

def sentenceTokens(text):
    text = text.lower()    # Convert the entire input text to lowercase
    return splitIntoSentences(text)  # Call the splitIntoSentences function 
                                     # on the lowercase and return the result as a list
def main():
    s = " I REALLY HOPE THIS WORKS. WHO KNOWS? LET'S SEE!"
    print(sentenceTokens(s))
main()
```

```python
import textlib
def sentenceCount(text):
    ## use sentenceTokens
    sentenceCounts = textlib.sentenceTokens(text)
    
    return len(sentenceCounts)   ## count the sentences

def main():
    s = " I really hope this works. Who knows? Let's see!"
    count = sentenceCount(s)
    print("Number of sentences:" , count)   ## print # of sentences, nice format
    
main()
```

```python
def averageWords(text):
    ## use wordCount and sentenceCount
    wordCounts = textlib.wordCount(text)
    sentenceCounts = textlib.sentenceCount(text)
    
    averageWordCount = wordCounts/sentenceCounts
    return averageWordCount
def main():
    s =  "Omg. I can't believe this. Will this work? I will never know. OMG I am nervous."  
    print(averageWords(s))
    
main()
```

```python
import matplotlib.pyplot as pyplot
import textlib

def plotIsPerParagraph(filename):
    paragraphNums = []   # Initialize a list to store the counts of "I" in each paragraph 
    IsInPara = 0    # Initialize a counter for the count of "I" in the current paragraph
    file = open(filename, 'r')  
    for line in file:
        if line.strip():  # Check if the line is not empty (non-blank)
            for word in line:
                if word == "I" or word == "I." or word == "I,": # Check if the word is "I", "I.", or "I,"  
                    IsInPara += textlib.wordCount("I")  # Count the instances of "I" using the textlib.wordCount function
        else:
            paragraphNums.append(IsInPara) # Append the count of "I" in the current paragraph to the list
            IsInPara = 0    # Reset the count for the next paragraph
            
    paragraphNums.append(IsInPara) # Append the count of "I" in the last paragraph
    file.close
    
    pyplot.plot(range(1, len(paragraphNums) + 1), paragraphNums)   
    pyplot.xlabel('Paragraph Number')
    pyplot.ylabel('Fraction of words that are "I"')
    pyplot.title('Count of "I" words per stanza In Creep by Radiohead')
    pyplot.show()

def main():
    plotIsPerParagraph('CreepLyrics.txt')  

main()
```
