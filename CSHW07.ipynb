## 5.5.15

```python
def daysAlive():
    """ Tells you how many days you've been alive for!
    
        Return Value: None
    """
    while True:
        try:
            text = input('How old are you?')
            age = int(text)
        except:                                                  ## if anything other than a num(int) is typed
            print('Please type ur age ONLY in numbers! Rude!!! >:( ') ## rude
        else:                                                             
            print('You have been alive for', round(age * 365.25), 'days! Yay!!')
            break                              ## don't want the code to keep prompting
                                               ## after giving the "correct" answer
def main():
    daysAlive()
    
main()
```

```python
def countCharacter(text, letter):
    """
    Return number of characters
    Parameters:
        text: a string
        letter: one character string
    
    Return value: number of occurances of letter in the text
    ( one character string in another string) 
    """
    
    
    charCount = 0                      ## initializing value to zero
    for char in text:                  ## for loop
        if char == letter:
            charCount = charCount + 1  ## accumlutator loop
    return charCount
    
def main():
    print(countCharacter("Omg let's hope this works", "t")) ## prints character count
    print(countCharacter("Omg let's hope this works", "o")) ## (There's 3 o's but counts 2 b/c of lower vs upper)
                                                        ## Question to consider: Is there a way to get a 
                                                        ## function to count upper and lower case of
main()                                                  ## letter as not two seperate characters?
```
