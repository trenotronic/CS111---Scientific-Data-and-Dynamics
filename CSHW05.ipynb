```python
def clockTime(ticks):
    """ Prints a second by second time of all ticks from midnight.
    
    Parameters:
        ticks: number of ticks
        
    Return Value: None
    """
    
    minutes = 0   #intializing both values to zero
    hours = 0
    for time in range(ticks):
        print('{0:>2}:{1:0>2}' .format(hours,minutes))  #to get answer to be in digital clock form
        minutes = (minutes +1) % 60
        if minutes < 1:                # once mintues count starts over again,
            hours = (hours + 1)        # if statement allows for hour to increase 
           
        
def main():
    clockTime(100)
    
main()
```

## 4.1.22

```python
def bacteria(intialPopSize, growthRate, days):
    """ Simulates a bacteria colony, and prints daily
    colony growth. 
    
    Parameter:
        days: number of days to simulate
        intialPopSize: intial population size
        growthRate: the rate of growth in population
        
    Return Value: the growth of the colony 
    """
    
    colony = intialPopSize
    for hours in range(days):
        hours = days * 24
        colony = (growthRate + 1) * hours
        
    return colony

def main():
    intialPopSize = int(input("Input the Intial Bacteria Population."))
    growthRate = float(input("Input the rate at which the bacteria grows(in decimal please!)."))
    finalColony = bacteria(intialPopSize, growthRate, 7)
    print('The colony after a week is ' + str(finalColony) + '.')
    
main()
```

```python
def sumDigits(number, numDigits):
    """
    
    Parameter:
        number: the input of number(s)
        numDigits: how many digits are in the number
        
    Return Value: the sum of the individual digits of number that has the number of digits
    """
    sumDigit = 0  #intializing value 
    for i in range(numDigits):
        sumDigit = (number//10**i) % 10 + sumDigit  #accumulating the values in loop
                
    return sumDigit

def main():
    answer = sumDigits(1234, 4)
    print(answer)
        
main()
```
