<!-- Explain the difference between an abstract data type and a python class -->

Without describing how they function specifically, abstract data type defines how to use a class of 'things'. ADT is made up of two componets: attributes for maintaining the data, and operations for accessing or changing the data. These attributes from ADT are transferred to a Python class, which keeps them in a set of instance variables for use in operations and makes them accessible to attributes via methods. The python class is what implements the abstract data type.

```python
nestDistance = 3
totalSteps = 300
stepLength = nestDistance/ totalSteps
```

```python
import turtle
ant = turtle.Turtle()
ant.pencolor('pink')
fly = turtle.Turtle()
fly.pencolor('green')
screen = ant.getscreen()
screen = fly.getscreen()
screen.setworldcoordinates(0, 0, 3, 1)
fly.goto(0, 1)
fly.dot()
ant.goto(0, 0)
ant.dot()

for count in range(300):
    angle = fly.towards(ant)
    fly.setheading(angle)
    fly.forward(stepLength)
    ant.forward(stepLength)
    
fly.distance(ant) 
finalDistance = fly.distance(ant)
print(finalDistance)
```
