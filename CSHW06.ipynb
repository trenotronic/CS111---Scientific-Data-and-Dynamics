```python
import matplotlib.pyplot as pyplot

def amortSimple(initAmount, r, payment, years):
    """Plots remaining balance over time for amortized loan.
    Parameters:
        initAmount (int)
        r = interest rate per year (float btwn 0 and 1)
        payment = monthly payment made (int or float)
        years (int)
    
    """
    
    A = [initAmount]                             ## initialize list of balances
    P = [payment]                                ## initialize list of payment
    R = [r/12]                                   ## initialize the rate
    tVals = [0]                                  ## initialize list payment times (months)
    for i in range(1, years*12+1):               ## iterating over each month
        interest = A[-1]*r/12                    ## calculating interest for current month
        A.append(A[-1] + interest - payment)     ## Adding current balance, after interest and payment, to the list
        P.append(P[-1] + payment)                ## Adding curent payment, to the list
        R.append(R[-1] + interest)               ## Adding current interest, to the list
        tVals.append(i)                          ## Adding current month to list
        
    pyplot.plot(tVals, A, color = 'red', label = 'Amount remaining')      ## plotting lists
    pyplot.plot(tVals, P, color = 'green', label = "Total Payment")
    pyplot.plot(tVals, R, color = 'blue', label = "Total Interest")
    
   
    leg = pyplot.legend(loc = "upper left")                      ## the little key
    pyplot.title('The coolest graph ever numba 1')               ## I never lie
    pyplot.xlabel('Months')                                      ## labeling axis'
    pyplot.ylabel('Amount of Money')
    pyplot.show()
    print(' The total payment is about '+ str(int(P[-1])) + ' and the total interest is about ' + str(int(R[-1])))

def main():
    amortSimple(165000, .0325, 1159.40, 15)
    
main()
```
