```python
import random
"""
    Determine if at least two people in a group share the same birthday.

    Parameters:
        numPeople (int): The number of people in the group.

    Return Value:
        bool: True if at least two people share a birthday, False otherwise.
"""
def sameBirthday(numPeople):

    birthdays = []         ## initialize list
    for birthday in range(numPeople):               ## Generate random birthdays for 
        birthdays.append(random.randint(0, 364+1))    ## the given number of people

    for i in range(numPeople):    ## Check for duplicates in the list of birthdays
        for j in range(i + 1, numPeople):
            if birthdays[i] == birthdays[j]:
                return True  ## Found a pair with the same birthday

    return False ## No duplicates, all birthdays are unique <3

    ## EXAMPLE
numPeople = 10
result = sameBirthday(numPeople)
print(f'In a group of {numPeople} people, at least two people share a birthday: {result}')

def birthdayProblem(numPeople, trials):
    """ Perform a Monte Carlo simulation to estimate the probability of at least two people
    sharing a birthday in a group.

    Parameters:
        numPeople (int): The number of people in the group.
        trials (int): The number of simulation trials to perform.

    Returns:
        float: The estimated probability of at least two people sharing a birthday.
    """   
    
    sharedBirthdays = 0   ## initialize

    for birthdays in range(trials):     ## Check if at least two people share a 
        if sameBirthday(numPeople):     ## birthday using the sameBirthday function
            sharedBirthdays += 1        ## add to count

    probability = sharedBirthdays / trials
    return probability

    ## EXAMPLE
numPeople = 10
numTrials = 15
result = birthdayProblem(numPeople, numTrials)
print(f'In a room with {numPeople} people, estimated probability of at least two people sharing a birthday: {result}')
```
