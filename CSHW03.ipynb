```python
import turtle
fitzgerald = turtle.Turtle()
fitzgerald.speed(5)
def drawPolygon(tortoise, sideLength, numSides):
        Angle = 180 - (numSides - 2) * (180/numSides)
        for i in range(numSides):
            tortoise.forward(sideLength)
            tortoise.left(Angle)
            
def polyFlower(tortoise, sideLength, numSides, numPolygons):
        for j in range(numPolygons):
            drawPolygon(tortoise, sideLength, numSides)
            tortoise.left(360/numPolygons)
        
polyFlower(fitzgerald, 40, 12, 7)
```

```python
import math

a = int(input("What is the value for a?"))
b = int(input("What is the value of b?"))
c = int(input("What is the value of c?"))
   
def quadratic(a,b,c):
    """ Uses the quadratic formula to solve for the two values of x
    Parameters:
        a: the value times x-squared
        b: the value times x
        c: constant
        
    Return value: x1, x2
    """ 
    inside = (b ** 2) - (4 * a * c)
    x1 = (-b - (math.sqrt(inside))) / (2 * a)
    x2 = (-b + (math.sqrt(inside))) / (2 * a)
    return x1, x2

def main():
    x1,x2 = quadratic(a,b,c)
    print(' The x-values are ' + str(x1) + ' and ' + str(x2) +'.')
    
main()
```
