## 6.3.9

```python
def capitalize(word):
    
    floWord = ord(word[0])  ## FLO -> first letter of word
                            ## Get the ASCII value of the first letter of the word
    if len(word) == 0:  ## Check if the word is not empty
        return word
    
    if 97 <= floWord <= 122:     ## Check if the first letter is a lowercase letter
        capfloWord = chr(floWord - 32)   # Convert the first letter to uppercase
        return capfloWord + word[1:]
    else:
        return word    ## If the first letter is not a lowercase letter, return the word unchanged
    
def main():
    ex1 = "Charlie"
    ex2 = "and"
    ex3 = "snoopy!"    ## Best duo
    
    print(capitalize(ex1))
    print(capitalize(ex2))
    print(capitalize(ex3))

main()
```

```python
def digit2String(digit):
    
     if 0 <= digit <= 9:              ## Convert the digit to its 
        return chr(ord('0') + digit)  ## corresponding ASCII character
     else:
        return None
    
def int2String(number):
    """
    Converts a positive integer to its string equivalent.

    Parameters:
        number (int): The positive integer to convert.

    Return value:
        str or None: The string equivalent of the integer, or None if any digit is not a decimal digit.
    """
    
    result = ''
    while number > 0:           ## Iterate through each digit of the number
        digit = number % 10       ## Extract the last digit
        digitString = digit2String(digit)    ## Convert the digit to its string representation
        result = digitString + result      ## Add the digit string to the result
        number = number//10             ## Remove the last digit from the number
        
        if digitString is None:
            return None  ## Return None if the digit is not a decimal digit
        
    return result

def main(): 
    print(int2String(1234))

main()
```

## 6.3.11

```python
import random

def randomPlate(length):
    """ Gives a randomly generating license plate in the example of 'ABC 123' format
    
    Parameters: 
        length: how long letters or number is on the plate
        
    Return value: a string representing a randomly generated license plate 
    """ 
    letters = ''                # initialize as string
    for i in range(length):               # create for loop
        letters += chr(random.randrange(65, 91))    # Generate random uppercase letters
        
    numbers = ''               # initialize as string
    for i in range(length):         
        numbers += str(random.randrange(10))   # Generate random digits
    licensePlate = letters + ' ' + numbers    # Combine letters, space, and numbers to form the license plate
    
    return licensePlate

def main():
    print(randomPlate(3))
    
main()
```
